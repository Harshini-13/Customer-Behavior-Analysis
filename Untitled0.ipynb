```python
import pandas as pd

df = pd.read_csv('/content/ecommerce_customer_data_large.csv')

print(df.head())
```

```python
print("Shape:", df.shape)

print("\nColumns:")
print(df.columns)

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
print(df.describe())
```

```python
print(df['Product Category'].value_counts())
```

```python
import matplotlib.pyplot as plt

df['Product Category'].value_counts().plot(kind='bar')

plt.title("Product Category Distribution")
plt.xlabel("Product Category")
plt.ylabel("Count")
plt.show()
```

```python
print(df['Gender'].value_counts())
```

```python
print(df['Churn'].value_counts())
```

```python
import matplotlib.pyplot as plt

df['Churn'].value_counts().plot(kind='bar')

plt.title("Customer Churn Distribution")
plt.xlabel("Churn")
plt.ylabel("Count")
plt.show()
```

```python
import matplotlib.pyplot as plt

df['Age'].hist(bins=10)

plt.title("Customer Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
from sklearn.cluster import KMeans
X = df[['Age', 'Total Purchase Amount']]

kmeans = KMeans(n_clusters=4, random_state=42)

df['Cluster'] = kmeans.fit_predict(X)

print(df['Cluster'].value_counts())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.scatter(df['Age'],
            df['Total Purchase Amount'],
            c=df['Cluster'])

plt.xlabel("Age")
plt.ylabel("Total Purchase Amount")
plt.title("Customer Segmentation")

plt.show()
```

```python
import matplotlib.pyplot as plt

df.groupby('Product Category')['Total Purchase Amount'].mean().plot(kind='bar')

plt.title("Average Purchase Amount by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Average Purchase Amount")
plt.show()
```

```python
import matplotlib.pyplot as plt

df.groupby('Gender')['Churn'].mean().plot(kind='bar')

plt.title("Churn Rate by Gender")
plt.xlabel("Gender")
plt.ylabel("Churn Rate")
plt.show()
```

```python
segment_summary = df.groupby('Cluster').agg({
    'Age':'mean',
    'Total Purchase Amount':'mean'
}).round(2)

print(segment_summary)
```
